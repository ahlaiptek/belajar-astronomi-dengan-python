```python
import math

# Konversi Qaus (busur) -> Jaib (sinus dalam ilmu falak)
def qaus_to_jaib(qaus):
    jaib = math.sin(math.radians(qaus)) * 60
    return jaib

print("Konversi Qaus ke Jaib")

qaus = 30
jaib = qaus_to_jaib(qaus)
print(f"{qaus} Qaus = {jaib:.2f} Jaib")

qaus = 60
jaib = qaus_to_jaib(qaus)
print(f"{qaus} Qaus = {jaib:.2f} Jaib")
```
