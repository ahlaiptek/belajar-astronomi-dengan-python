```python
# Buruj (برج) = Rasi bintang/Zodiak (1-12 atau 0-11)
# Darojah (درجة) = Derajat (0-29 dalam satu buruj)
# Daqiqah (دقيقة) = Menit busur (0-59)
# Tsawani (ثوان) = Detik busur (0-59)
# Tsawalits (ثوالث) = Pertiga detik (0-59)
```

```python
class PositisiAstronomi:
    """Posisi benda langit dalam zodiak (untuk hisab)"""
    
    BURUJ_NAMES = [
        "Haml", "Tsaur", "Jauza", "Saratan", "Asad", "Sunbulah",
        "Mizan", "Aqrab", "Qaus", "Jadyu", "Dalwu", "Hut"
    ]
    
    def __init__(self, buruj, darojah, daqiqah, tsawani, tsawalits):
        self.buruj = buruj  # 1-12 atau 0-11
        self.darojah = darojah
        self.daqiqah = daqiqah
        self.tsawani = tsawani
        self.tsawalits = tsawalits
    
    def __str__(self):
        nama_buruj = self.BURUJ_NAMES[self.buruj - 1]
        return f"{nama_buruj} {self.darojah}° {self.daqiqah}′ {self.tsawani}″ {self.tsawalits}‴"
    
    def ke_desimal(self):
        """Konversi ke derajat desimal dari titik Aries (0°)"""
        total_derajat = (
            (self.buruj - 1) * 30 +  # Setiap buruj = 30°
            self.darojah +
            self.daqiqah / 60 +
            self.tsawani / 3600 +
            self.tsawalits / 216000
        )
        return total_derajat

# Contoh: Posisi Matahari di Asad 15° 26′ 13″ 30‴
matahari = PositisiAstronomi(5, 15, 26, 13, 30)
print(matahari)  # Asad 15° 26′ 13″ 30‴
print(f"Dari titik Aries: {matahari.ke_desimal()}°")
```

```python
def penjumlahan_tsawani_tsawalits(tsawani_a, tsawalits_a, tsawani_b, tsawalits_b):
    """
    Penjumlahan tsawani dan tsawalits dengan carry over
    
    Args:
        tsawani_a, tsawalits_a: Nilai pertama
        tsawani_b, tsawalits_b: Nilai kedua
    
    Returns:
        tuple: (tsawani_hasil, tsawalits_hasil)
    """
    # Penjumlahan tsawalits
    total_tsawalits = tsawalits_a + tsawalits_b
    
    # Carry over ke tsawani (1 tsawani = 60 tsawalits)
    carry_tsawani = total_tsawalits // 60
    sisa_tsawalits = total_tsawalits % 60
    
    # Penjumlahan tsawani + carry
    total_tsawani = tsawani_a + tsawani_b + carry_tsawani
    
    # Carry over ke daqiqah jika perlu (1 daqiqah = 60 tsawani)
    carry_daqiqah = total_tsawani // 60
    sisa_tsawani = total_tsawani % 60
    
    return sisa_tsawani, sisa_tsawalits, carry_daqiqah


# Contoh penggunaan
tsawani1, tsawalits1 = 1, 45
tsawani2, tsawalits2 = 1, 15

hasil_tsawani, hasil_tsawalits, carry = penjumlahan_tsawani_tsawalits(
    tsawani1, tsawalits1, tsawani2, tsawalits2
)

print(f"{tsawani1}″ {tsawalits1}‴ + {tsawani2}″ {tsawalits2}‴")
print(f"= {hasil_tsawani}″ {hasil_tsawalits}‴ + {carry} (carry ke daqiqah)")
```
